## Preprocessor

```python
# Preprocessor
class Preprocessor(BaseEstimator, TransformerMixin):


    def __init__(self, lower=True, strip=True, remove_punct=True, remove_diacritics=True, unicode_form='NFC', remove_stop=False, stem_tokens=False, lemmatize_tokens=True, language='english'):
        
        self.lower      = lower
        self.strip      = strip
        self.remove_punct = remove_punct
        self.remove_diacritics = remove_diacritics
        self.set_unicode_form(unicode_form)
     
        self.remove_stop = remove_stop
        self.stem_tokens = stem_tokens
        self.lemmatize_tokens = lemmatize_tokens    

        self.stopwords  = set(nltk.corpus.stopwords.words(language))
        self.punct      = set(string.punctuation)

        self.lemmatizer = WordNetLemmatizer()
        self.stemmer = nltk.stem.SnowballStemmer(language) 
        self.tokenizer = TweetTokenizer()

 
    def transform(self, X):
        return [
            list(self.tokenize(doc)) for doc in X
        ]

    def inverse_transform(self, X):
        return [" ".join(doc) for doc in X]


    def fit(self, X, y=None):
        return self

 
    def set_stopwords(self, stopwords):
        self.stopwords  = stopwords or set(nltk.corpus.stopwords.words('english'))


    def set_punct(self, punct):
        self.punct      = punct or set(string.punctuation)


    def set_unicode_form(self, unicode_form):
        self.unicode_form = unicode_form.strip().upper()
        if not self.unicode_form in ['NFC','NFD','NFKC','NFKD']:
            self.unicode_form = 'NFC'


    def show_settings(self):
        print("lower: ",self.lower)
        print("strip: ",self.strip)
        print("remove_punct: ",self.remove_punct)
        print("remove_diacritics: ",self.remove_diacritics)
        print("unicode_form: ",self.unicode_form)
        print("remove_stop: ",self.remove_stop)
        print("stem_tokens: ",self.stem_tokens)
        print("lemmatize_tokens",self.lemmatize_tokens)       
        print("lemmatizer: ",type(self.lemmatizer).__name__)
        print("stemmer: ",type(self.stemmer).__name__) 
        print("punctuation: ",self.punct)
        print("stopwords: ",self.stopwords)


    # convert nltk pos-tag to wordnet pos-tag for use by the wordnet lemmatizer
    @staticmethod
    def get_wordnet_tag(nltk_tag):
        if nltk_tag.startswith('N'):
            return wn.NOUN
        elif nltk_tag.startswith('V'):
            return wn.VERB
        elif nltk_tag.startswith('J'):
            return wn.ADJ
        elif nltk_tag.startswith('R'):
            return wn.ADV
        else:          
            return None

    # determine whether a string is comprised of all punctuation-like characters
    def is_punct(self, text):
        if text in self.punctuation:
            return True
        if regex.match(r"[\p{P}\p{Mn}\p{Sk}]+", text):
            return True
        return False


    # normalize token strings
    def normalize_string(self, token):

        token = html.unescape(token)
        token = token.lower() if self.lower else token
        token = token.strip(' _*')  if self.strip else token

        if self.remove_diacritics:
            token = regex.sub("\p{Mn}",'',unicodedata.normalize('NFD',token)) 

        #not is in the stopword list, but n't isn'
        if token == "n't" and self.stopwords:
            token = "not"
      
        return unicodedata.normalize(self.unicode_form,token)

    # determine whether a string is comprised completely of things that seem like punctuation
    def is_punct(self, text):
        if text in string.punctuation:
            return True
        if regex.match(r"^[\p{P}\p{Mn}\p{Sk}]+$", text):
            return True
        return False

    # tokenize the document with optional normalization
    def tokenize(self, document, all_fields=False):
        tokens = []
        for sent in sent_tokenize(document):
            cleaner_sent = html.unescape(sent)
            for token in nltk.pos_tag(self.tokenizer.tokenize(cleaner_sent)):
                stem = ''
                token_text = self.normalize_string(token[0])
                token_pos = token[1]
                if self.remove_punct and self.is_punct(token_text):
                #if remove_punct and token_text in string.punctuation:
                    continue
                if self.remove_stop and token_text in self.stopwords:
                    continue
                if self.stem_tokens or all_fields:
                    stem = self.stemmer.stem(token_text)
                if self.lemmatize_tokens or all_fields:
                    wordnet_tag = self.get_wordnet_tag(token_pos)
                    if wordnet_tag is not None:
                        lemma = self.lemmatizer.lemmatize(token_text,wordnet_tag)
                    else:
                        lemma = token_text
                if all_fields:
                    tokens.append({'token': token_text, 'stem': stem, 'lemma': lemma})
                elif self.stem_tokens:
                    tokens.append(stem)
                elif self.lemmatize_tokens:
                    tokens.append(lemma)
                else:
                    tokens.append(token_text)     
        return tokens
```

```python
vocabulary = {}
user_vocabularies = {}
my_preprocessor = Preprocessor(remove_punct=True, remove_stop=False, remove_diacritics=True, lemmatize_tokens=True)


# create a method for doing token analysis and generating the counts we need
def get_token_analysis(tweet_text, user=None):
    tokens = my_preprocessor.tokenize(tweet_text)
    for token in tokens:
            if token in vocabulary:
                vocabulary[token] += 1
            else:
                vocabulary[token] = 1
            if user:
                if user in user_vocabularies:
                    user_dict = user_vocabularies[user]
                    if token in user_dict:
                        user_dict[token] += 1
                    else:
                        user_dict[token] = 1
                    user_vocabularies[user] = user_dict
                else:
                    user_vocabularies[user] = {token: 1}
            
    return tokens
```

```python
# open the input data and print out basic file information
data_file = './CleanedTweets.csv'
input_df = pd.read_csv(data_file,sep=',')
print("headers: ",list(input_df.columns)) # print columns names
print("rows: ",len(input_df), "\n") # print number of rows/observations (number of Tweets)
print(input_df['UserName'].value_counts()) # Print each of the users and tweet counts
```

```python
# set up some variables to hold total counts
# note: some of these are extraneous because you can get them from the data frame
# however, we can use them to double check our calculations
num_tweets = 0
total_tokens = 0
total_tweet_length_in_chars = 0
tweet_lengths_in_tokens = []

# iterate over the rows in your dataframe, keeping track of various counts
for row in input_df.itertuples():
    num_tweets += 1
    total_tweet_length_in_chars += len(row.FullText)
    tokens = get_token_analysis(row.FullText, row.UserName)
    tweet_lengths_in_tokens.append(len(tokens))
    total_tokens += len(tokens)

# reverse frequency sort the vocabulary
sorted_vocabulary = sorted(vocabulary.items(), key=lambda x:x[1], reverse=True)

# reverse frequency sort all of the user-specific dictionaries
for current_user in user_vocabularies.keys():
    cur_dict = user_vocabularies[current_user]
    sorted_dict = sorted(cur_dict.items(), key=lambda x:x[1], reverse=True)
    user_vocabularies[current_user] = sorted_dict

# add the token count for each tweet to the dataframe 
input_df['LengthInTokens'] = tweet_lengths_in_tokens
print("headers: ",list(input_df.columns))
```

# Tweet Part of Speech Analysis by UserName

```python
import spacy
from spacy import displacy
from spacy.matcher import Matcher
from spacy.matcher import PhraseMatcher
from spacy.symbols import ORTH
from spacy.util import minibatch, compounding


from itertools import chain
import matplotlib.pyplot as plt 
from matplotlib.ticker import MaxNLocator
import numpy as np
import random
import sys
import time
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
# collect POS information for all tweets for each user
pos_Klobuchar = []
pos_Sanders = []
pos_Warren = []
pos_Biden = []
pos_Buttigieg = []
pos_Steyer = []
pos_Gabbard = []

Klobuchar_tweets = input_df[input_df['UserName'] == 'Amy Klobuchar']['FullText']
for tweet in Klobuchar_tweets:
    cur = nlp(tweet)
    for token in cur:
        pos_Klobuchar.append(token.pos_)

Sanders_tweets = input_df[input_df['UserName'] == 'Bernie Sanders']['FullText']
for tweet in Sanders_tweets:
    cur = nlp(tweet)
    for token in cur:
        pos_Sanders.append(token.pos_)


Warren_tweets = input_df[input_df['UserName'] == 'Elizabeth Warren']['FullText']
for tweet in Warren_tweets:
    cur = nlp(tweet)
    for token in cur:
        pos_Warren.append(token.pos_)
        
Biden_tweets = input_df[input_df['UserName'] == 'Joe Biden']['FullText']
for tweet in Biden_tweets:
    cur = nlp(tweet)
    for token in cur:
        pos_Biden.append(token.pos_)
        
Buttigieg_tweets = input_df[input_df['UserName'] == 'Pete Buttigieg']['FullText']
for tweet in Buttigieg_tweets:
    cur = nlp(tweet)
    for token in cur:
        pos_Buttigieg.append(token.pos_)
        
Steyer_tweets = input_df[input_df['UserName'] == 'Tom Steyer']['FullText']
for tweet in Steyer_tweets:
    cur = nlp(tweet)
    for token in cur:
        pos_Steyer.append(token.pos_)
        
Gabbard_tweets = input_df[input_df['UserName'] == 'Tulsi Gabbard ðŸŒº']['FullText']
for tweet in Gabbard_tweets:
    cur = nlp(tweet)
    for token in cur:
        pos_Gabbard.append(token.pos_)
```

```python
# get Part of Speech counts for each UserName
from collections import Counter

pos_counts_Klobuchar = Counter(pos_Klobuchar)
pos_counts_Sanders = Counter(pos_Sanders)
pos_counts_Warren = Counter(pos_Warren) 
pos_counts_Biden = Counter(pos_Biden)  
pos_counts_Buttigieg = Counter(pos_Buttigieg) 
pos_counts_Steyer = Counter(pos_Steyer) 
pos_counts_Gabbard = Counter(pos_Gabbard)
```

```python
# define function to calculate get POS percentages from POS counts in UserName dictionaries
def div_d(my_dict):
    sum_p = sum(my_dict.values())
    for i in my_dict:
        my_dict[i] = float(my_dict[i]/sum_p)
    return my_dict 
```

```python
# create POS useage by UserName dataframe 
df1 = pd.DataFrame(list(div_d(pos_counts_Klobuchar).items()),columns = ['POS',''])
df2 = pd.DataFrame(list(div_d(pos_counts_Sanders).items()),columns = ['POS','']) 
df3 = pd.DataFrame(list(div_d(pos_counts_Warren).items()),columns = ['POS','Warren']) 
df4 = pd.DataFrame(list(div_d(pos_counts_Biden).items()),columns = ['POS','Biden']) 
df5 = pd.DataFrame(list(div_d(pos_counts_Buttigieg).items()),columns = ['POS','Buttigieg']) 
df6 = pd.DataFrame(list(div_d(pos_counts_Steyer).items()),columns = ['POS','Steyer']) 
df7 = pd.DataFrame(list(div_d(pos_counts_Gabbard).items()),columns = ['POS','Gabbard']) 

df = df1.merge(df2, on = 'POS', suffixes=('Klobuchar','Sanders'))
df = df.merge(df3, on = 'POS', suffixes=('','Warren'))
df = df.merge(df4, on = 'POS', suffixes=('','Biden'))
df = df.merge(df5, on = 'POS', suffixes=('','Buttigieg'))
df = df.merge(df6, on = 'POS', suffixes=('','Steyer'))
df = df.merge(df7, on = 'POS', suffixes=('','Gabbard'))
df
```

```python
#sns.set()
df.plot(kind='bar', stacked=True, legend = False)
```

```python
dic1 = pos_counts_Klobuchar
dic2 = pos_counts_Sanders
dic3 = pos_counts_Warren
dic4 = pos_counts_Biden
dic5 = pos_counts_Buttigieg
dic6 = pos_counts_Steyer
dic7 = pos_counts_Gabbard

result = {}
for key in (dic1.keys() | dic2.keys()):
    if key in dic1: result.setdefault(key, []).append(dic1[key])
    if key in dic2: result.setdefault(key, []).append(dic2[key])
    if key in dic3: result.setdefault(key, []).append(dic3[key])
    if key in dic4: result.setdefault(key, []).append(dic4[key])
    if key in dic5: result.setdefault(key, []).append(dic5[key])
    if key in dic6: result.setdefault(key, []).append(dic6[key])
    if key in dic7: result.setdefault(key, []).append(dic7[key])

#print(result)
```
